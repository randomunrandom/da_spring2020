```python
Sa = np.sqrt(stats.tvar(a))
Sb = np.sqrt(stats.tvar(b))
Sa, Sb
```

```python
np.cov(np.vstack([a, b]))
```

```python
c = np.cov(np.vstack([a, b]))[0,1]
c
```

```python
c / (Sa * Sb)
```

```python
cov_(a * b).sum() / (a.size - 1) - a.mean() * b.mean()
```

```python
covar_matrix =  np.cov(np.vstack([a, b]))
r = covar_matrix[0, 1] / np.sqrt(covar_matrix[0, 0] * covar_matrix[1, 1])
r
```

```python
T = np.sqrt(a.size - 2) * r / np.sqrt(1 - r**2)
T
```

```python
stats.pearsonr(a, b)
```

```python
2 * (1 - stats.t.cdf(T, a.size - 2))
```

# Семинар

```python
a = stats.norm.rvs(0, 1, size=30)
a
```

```python
b = a * 1.35 + stats.norm.rvs(0, 0.6, size=30)
b
```

```python
sns.scatterplot(a, b)
plt.show()
```

```python
c = np.cov(np.vstack([a, b]))
c
```

```python
r = c[0, 1] / np.sqrt(c[0, 0] * c[1, 1])
r
```

```python
T = np.sqrt(a.size - 2) * r / np.sqrt(1 - r**2)
T
```

```python
1 - stats.t.cdf(T, a.size - 2)
```

```python
2*(1 - stats.t.cdf(T, a.size - 2))
```

```python
stats.pearsonr(a, b)
```

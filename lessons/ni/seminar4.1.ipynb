```python
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns; sns.set()
```

```python
x = np.array([
    10043, 9596, 10305, 8354, 9413, 19776, 9815, 11311, 11253, 10856, 11389
])

y = np.array([
    14253, 7843, 9581, 8594, 16119, 10112, 10173, 9756
])
```

```python
sns.distplot(x)
```

```python
sns.distplot(y)
```

```python
print(x.mean())
print(y.mean())
print(stats.tvar(x))
print(stats.tvar(y))
```

```python
print(stats.shapiro(x))
print(stats.shapiro(y))
```

# Student

```python
# H1: y < x
# 1. T-statistic < 0
s, p = stats.ttest_ind(y, x)
s, p
print(p/2)
```
